```python
#create a dataframe w a column called Tweets
#df=tweetData[['Tweets',axis=1]]
df=tweetData['Tweet']
Sdf=tweetData['Sentiment']


#show the first 5 rows
df.head()
Sdf.head()
```

```python
#Clean the text

#create a function to clean the tweets
def cleanTxt(text):
  text=re.sub(r'@[A-Za-z0-9]+', '',text) #remeoves @mentions
  text=re.sub(r'#','',text) #remove the hashtag symbol
  text=re.sub(r'RT[\s]+', '', text) #remove retweets 
  text=re.sub(r'https?:\/\/\S+', '', text) #remove hyperlink

  return text 

#cleaning text  
#df['Tweets']= df['Tweets'].apply(cleanTxt)  
df=df.apply(cleanTxt)  
#show cleaned text
df

```

```python
merged = pd.concat([Sdf, df], axis=1)
```

```python
#cleanedData.csv=df.to_csv()
merged.to_csv(r'cleanedData.csv')

from google.colab import files
files.download("cleanedData.csv")
```
