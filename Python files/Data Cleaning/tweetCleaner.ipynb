```python
import pandas as pd
import numpy as np
import re
from google.colab import drive 
drive.mount('/content/drive')
```

```python
header = ["target", "id", "date", "flag", "user", "text"]

tw= pd.read_csv(r"/content/drive/MyDrive/WPI/CS 539 Final Project/Tweet data/training.1600000.processed.noemoticon.csv", encoding='latin-1',names = header)
tw
```

```python
#Clean the text

#create a function to clean the tweets
def cleanTxt(text):
  text=re.sub(r'@[A-Za-z0-9]+', '',text) #remeoves @mentions
  text=re.sub(r'#','',text) #remove the hashtag symbol
  text=re.sub(r'RT[\s]+', '', text) #remove retweets 
  text=re.sub(r'https?:\/\/\S+', '', text) #remove hyperlink

  return text 

#cleaning text  
tw['text']= tw['text'].apply(cleanTxt)  

#show cleaned text
tw
```

```python
#Bound target as 0 for negative and 1 for positive
tw["target"]= tw["target"]/4
tw
```
